```python
import pandas as pd
import numpy as np
import seaborn as sns # Used here to easily load the Titanic dataset

# 1. Load Titanic Dataset (via Seaborn library)
df_titanic = sns.load_dataset('titanic')

print("Titanic Dataset Loaded Successfully")
```

```python
# Check Data Types and Null Values
print("--- Info & Missing Values ---")
df_titanic.info()

# Statistical Summary for Numerical Columns
print("\n--- Numerical Summary ---")
display(df_titanic.describe())

# Statistical Summary for Categorical Columns (include='O')
print("\n--- Categorical Summary ---")
display(df_titanic.describe(include=['O', 'category']))
```

```python
# Check unique values in a specific categorical column
print("Unique Embark Towns:", df_titanic['embark_town'].unique())
print("Unique Classes:", df_titanic['class'].unique())

# Check for Data Imbalance in Target Variable
print("\n--- Target Variable Distribution (Survived) ---")
print(df_titanic['survived'].value_counts(normalize=True) * 100)
```

## Project: Titanic Dataset Exploratory Analysis

## 1. Data Structure Overview
* **Dataset Size:** The dataset contains **891 rows** and **15 columns**.
* **Suitability:** This size is considered small but sufficient for learning basic Machine Learning (ML) classification algorithms.

## 2. Variable Identification
Based on `df.head()` and manual inspection, the features can be classified as:

* **Numerical:** `age`, `fare`, `sibsp` (Siblings/Spouses), `parch` (Parents/Children).
* **Categorical (Nominal):** `sex`, `embarked`, `who`, `deck`.
* **Categorical (Ordinal):** `class` (First > Second > Third), `pclass`.
* **Binary:** `adult_male` (True/False), `alone` (True/False).
* **Target Variable:** `survived` (0 = No, 1 = Yes). This makes it a **Classification** problem.

## 3. Data Quality & Observations

### Missing Values:
* The `deck` column has a massive number of missing values (approx **77%**). It may need to be dropped.
* The `age` column has roughly **20%** missing values. We will need to fill these (imputation) before ML modeling.

### Imbalance:
* The target variable `survived` shows that roughly **61% died** and **38% survived**. This is a slight imbalance but acceptable for standard training without heavy modification.

## 4. ML Readiness Conclusion
The dataset is **Partially Ready**.

* **Action Required:** We must handle the missing `age` data and convert categorical text data (like 'male'/'female') into numbers (0/1) before feeding it into a model.
